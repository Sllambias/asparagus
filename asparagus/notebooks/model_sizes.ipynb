```python
from asparagus.modules.networks.nets import primus_g

# Initialize the model
model = primus_g(input_channels=1, output_channels=1, patch_size=(128, 128, 128))

# Count the number of parameters
num_params = sum(p.numel() for p in model.parameters() if p.requires_grad)

print(f"model has {num_params} parameters")
params_in_millions = num_params / 1_000_000
print(f"model has {params_in_millions:.2f}M parameters")

# count encoder params
num_params = sum(p.numel() for p in model.encoder.parameters() if p.requires_grad)

print(f"encoder has {num_params} parameters")
params_in_millions = num_params / 1_000_000
print(f"encoder has {params_in_millions:.2f}M parameters")
```

```python
from asparagus.modules.networks.nets import resenc_unet_b

# Initialize the model
model = resenc_unet_b(
    dimensions="3D",
    input_channels=1,
    output_channels=1,
    features_per_stage=(64, 128, 256, 512, 620, 620),
    n_blocks_per_stage_encoder=(1, 3, 4, 6, 6, 6),
    n_conv_per_stage_decoder=(1, 1, 1, 1, 1),
)

# Count the number of parameters
num_params = sum(p.numel() for p in model.parameters() if p.requires_grad)

print(f"model has {num_params} parameters")
params_in_millions = num_params / 1_000_000
print(f"model has {params_in_millions:.2f}M parameters")


# count encoder params
num_params = sum(p.numel() for p in model.encoder.parameters() if p.requires_grad)

print(f"encoder has {num_params} parameters")
params_in_millions = num_params / 1_000_000
print(f"encoder has {params_in_millions:.2f}M parameters")
```
